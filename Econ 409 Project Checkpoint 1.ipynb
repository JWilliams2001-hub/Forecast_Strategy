### Possible Choices

- **Currencies**
    - Models we've learned in class: Use Interest Rate Parity, Inflation Differentials, and Taylor Rule.
    - Fundamental: We can pull the Consumer Price Index (CPI) or Central Bank Policy Rates for two different countries (e.g., USD vs. EUR).
    - Conceptual Framework: We can argue that higher relative interest rates in one country often lead to currency appreciation as investors seek higher yields.
    - Data Source: Bloomberg Terminal for exchange rates and World Bank or FRED for macro data.
      
- **Equities (Stocks and S&P500 Sectors)**
    - Models we've learned in class: Quasi-Fischer, Kalmen Filter, differentiate "Value" or "Growth" stocks to forecast price movements.
    - Fundamental: We can pull 12-month forward EPS (Earnings Per Share) or P/E Ratios from Bloomberg.
    - Conceptual Framework: We can argue that when the forward EPS growth accelerates relative to the price, the stock is undervalued and likely to rise.
    - Data Source: Bloomberg Terminal
      
- **Commodities**
    - Models we've learned in class: Use Sentiment Analysis or Inflation Hedging.
    - Fundamental: Use Google Trends data for "inflation" or Gold ETF (GLD) flows from Bloomberg.
    - Conceptual Framework: Gold often acts as a contrarian asset; high retail "fear" or sentiment index peaks might signal an entry point. ]
    - Data Source: Google Trends and Bloomberg Terminal
